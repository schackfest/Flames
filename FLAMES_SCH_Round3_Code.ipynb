```python
# tạo 1 dictionary gồm tháng có demand forecast & tổng số ngày trong tháng đó (bao gồm Chủ Nhật)
date_list = []
eomonth_list = []


for ind,data in enumerate(df_demand_dnb.columns):
    date_list.append(datetime.strptime(data,'%b_%Y')) # append column heading as date into a list
    
    eomonth_list.append(calendar.monthrange(datetime.strptime(data,'%b_%Y').year,datetime.strptime(data,'%b_%Y').month)[1]) # find the last day of each month

month_eomonth = {date_list[i]:eomonth_list[i] for i in range(len(date_list))} # create a list of each month & its last day

# tạo 1 list ngày làm việc của Festory theo từng tháng (ngày trong tháng trừ Chủ Nhật)
date_exclude_Sun = []
for date,eomonth in month_eomonth.items():
    day = 1 # gán biến chạy là ngày đầu tiên hàng tháng
    while day <= eomonth: # make sure ngày có trong tháng 
        if datetime(date.year,date.month,day).weekday() != 6: # filter ngày chủ nhật
            date_exclude_Sun.append(datetime(date.year,date.month,day)) # add các ngày thỏa điều kiện vào 1 list
        day += 1
    continue

# tính TỔNG số ngày làm việc (working_day) của Festory theo từng tháng (ngày trong tháng trừ Chủ Nhật)

total_apr = total_may = total_jun = 0

for i in date_exclude_Sun:
    if i.month == 4:
        total_apr += 1 
    elif i.month == 5:
        total_may += 1
    else:
        total_jun += 1

total_day = [total_apr,total_may,total_jun]
working_day = {date_list[i].month:total_day[i] for i in range(len(date_list))}

# round up Daily_Demand to one decimal
def round_up(num):
    return math.ceil(num * 10) / 10

date_month = []
last_day = []

for d, l in month_eomonth.items():
    date_month.append(d.month)
    last_day.append(l)
    
month_n_last_day = {date_month[i]:last_day[i] for i in range(len(date_month))}
```

```python
#THAM KHẢO CASE FIFO, THÊM YẾU TỐ SAFETY STOCK
from datetime import datetime, timedelta
def xu_ly_SL_CONLAI(df_in):
  from datetime import datetime, timedelta
  global df_soh
  global df_imr
  global df_order
  global df_sog
  global dd 
  global safety_stock
  max_ind = df_in['THIẾU'].lt(0).idxmax() #10
  df_soh.loc[max_ind,'TỒN KHO CUỐI NGÀY'] = - (df_in.loc[max_ind,'THIẾU'])
  for ind, data in df_in.iterrows():
    
    qty_thieu = df_in.loc[ind,'THIẾU']

    #Nếu SL thiếu >= 0  ===> Lấy hết => Stock về 0 
    if qty_thieu >=0:
      df_soh.loc[ind, "TỒN KHO CUỐI NGÀY"]=0
  
  #Nếu xảy ra tình trạng hết hàng giữa chừng, ta tiến hành đặt bổ sung trong tháng 
  if qty_thieu > 0 :
    x = df_in.loc[ind,"NGUYÊN LIỆU"]
    ind = df_imr["NGUYÊN LIỆU"].eq(x).idxmax()
    moq = df_imr.loc[ind,"MOQ"]
    order = qty_thieu            
    rorder = math.ceil(order/moq)
    from datetime import datetime, timedelta
    df_imr["SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)"] = df_imr["SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)"].astype(float)
    HSD = dd + timedelta(days=df_imr.loc[ind,'SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)'])
    df_sog = df_sog.append({"NGUYÊN LIỆU":x,"SỐ LƯỢNG":rorder*moq-qty_thieu, "NGÀY VỀ":dd,"ĐƠN VỊ":df_imr.loc[ind,"ĐƠN VỊ"], "HẠN SỬ DỤNG":HSD},ignore_index=True)  
    df_imr["LEADTIME ĐẶT HÀNG (NGÀY)"] = df_imr["LEADTIME ĐẶT HÀNG (NGÀY)"].astype(float)
    df_order = df_order.append({"NGUYÊN LIỆU":x,"NHÀ CUNG CẤP":df_imr.loc[ind,"NHÀ CUNG CẤP"], "SỐ LƯỢNG":rorder, 
                     "NGÀY ĐẶT": dd - timedelta(days=df_imr.loc[ind,"LEADTIME ĐẶT HÀNG (NGÀY)"])}, ignore_index=True)
    exit
  
  #Lượng tồn kho trong kho tuột xuống dưới safety_stock cho 10 ngày --> tiến hành đặt bù 
  ind = df_in['inv_cumsum'].idxmax()
  if df_in.loc[ind, 'inv_cumsum'] < df_in.loc[ind, "NHU CẦU NGUYÊN LIỆU"]*10:
    order = df_in.loc[ind, "NHU CẦU NGUYÊN LIỆU"]*10 - df_in.loc[ind, 'inv_cumsum']       
    x = df_in.loc[ind,"NGUYÊN LIỆU"]
    ind = df_imr["NGUYÊN LIỆU"].eq(x).idxmax()
    moq = df_imr.loc[ind,"MOQ"]
    rorder = math.ceil(order/moq)
    df_imr["SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)"] = df_imr["SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)"].astype(float)
    HSD = dd + timedelta(days=df_imr.loc[ind,'SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)'])
    df_sog = df_sog.append({"NGUYÊN LIỆU":x,"SỐ LƯỢNG":rorder*moq-qty_thieu, "NGÀY VỀ":dd,"ĐƠN VỊ":df_imr.loc[ind,"ĐƠN VỊ"], "HẠN SỬ DỤNG":HSD},ignore_index=True)  
    df_imr["LEADTIME ĐẶT HÀNG (NGÀY)"] = df_imr["LEADTIME ĐẶT HÀNG (NGÀY)"].astype(float)
    df_order = df_order.append({"NGUYÊN LIỆU":x,"NHÀ CUNG CẤP":df_imr.loc[ind,"NHÀ CUNG CẤP"], "SỐ LƯỢNG":rorder, 
                     "NGÀY ĐẶT": dd - timedelta(days=df_imr.loc[ind,"LEADTIME ĐẶT HÀNG (NGÀY)"])}, ignore_index=True)
    exit
          
def day_by_day(df_in):
    from datetime import datetime, timedelta
    global df_soh
    global df_sog
    global df_bom
    global df_imr
    global df_order
    global df_expired
    global dd 
      
    df_sohh=pd.DataFrame()
    #gán mốc là ngày trong file  
    df_in = df_in.reset_index()
    dd = df_in.loc[0,"DATE"]
    #Nếu đây là đầu tháng
    if dd.day == 1:
        df_ofm = df_in.merge(df_bom, on = "THÀNH PHẨM", how = 'inner')
        df_ofm['MONTHLY_DEMAND_QTY']= df_ofm['MONTHLY_DEMAND_QTY'].astype(float)
        df_ofm['LƯỢNG NGUYÊN LIỆU CẦN']=df_ofm["MONTHLY_DEMAND_QTY"]*df_ofm["SỐ LƯỢNG NGUYÊN LIỆU TRÊN 1 THÀNH PHẨM"]*1.3
        df_ofm = df_ofm.groupby("NGUYÊN LIỆU").sum()
        df_sohh = df_soh.groupby("NGUYÊN LIỆU")["TỒN KHO CUỐI NGÀY"].sum()
        df_ofm = df_ofm.merge(df_sohh,on="NGUYÊN LIỆU",how='inner')
        df_ofm['CẦN ĐẶT']=df_ofm['LƯỢNG NGUYÊN LIỆU CẦN']- df_ofm['TỒN KHO CUỐI NGÀY']
        df_ofm=df_ofm.merge(df_imr,on = "NGUYÊN LIỆU", how = 'inner')
        df_ofm["LEADTIME ĐẶT HÀNG (NGÀY)"] = df_ofm["LEADTIME ĐẶT HÀNG (NGÀY)"].astype(float)
        df_ofm["SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)"] = df_ofm["SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)"].astype(float)
        for ind, data in df_ofm.iterrows():
            if df_ofm.loc[ind,"CẦN ĐẶT"] > 0:
                df_ofm.loc[ind,"ĐẶT THỰC TẾ"] = math.ceil(df_ofm.loc[ind,"CẦN ĐẶT"]/df_ofm.loc[ind,"MOQ"])

                df_sog = df_sog.append({"NGUYÊN LIỆU":df_ofm.loc[ind,"NGUYÊN LIỆU"],
                                            "SỐ LƯỢNG":df_ofm.loc[ind,"MOQ"]*df_ofm.loc[ind,"ĐẶT THỰC TẾ"],
                                            "NGÀY VỀ":dd,
                                            "HẠN SỬ DỤNG":dd + timedelta(days=df_ofm.loc[ind,"SHELF LIFE ĐƯỢC ƯỚC TÍNH KỂ TỪ NGÀY NHẬP KHO (NGÀY)"])
                                            ,"ĐƠN VỊ":df_ofm.loc[ind,"ĐƠN VỊ"]},ignore_index=True)
                df_order = df_order.append({"NGUYÊN LIỆU":df_ofm.loc[ind,"NGUYÊN LIỆU"],
                                            "NHÀ CUNG CẤP":df_ofm.loc[ind,"NHÀ CUNG CẤP"],
                                            "SỐ LƯỢNG":df_ofm.loc[ind,"ĐẶT THỰC TẾ"],
                                            "NGÀY ĐẶT": dd - timedelta(days=df_ofm.loc[ind,"LEADTIME ĐẶT HÀNG (NGÀY)"])}, ignore_index=True)
      
            else:
                df_ofm.loc[ind,"ĐẶT THỰC TẾ"] = 0
    
    # Kiểm tra hết hạn
    for ind, data in df_soh.iterrows():
        if df_soh.loc[ind,"HẠN SỬ DỤNG"] <= dd:
            # print(df_soh)
            df_soh["TỒN KHO CUỐI NGÀY"].fillna(0,inplace=True)
            df_soh["TỒN KHO CUỐI NGÀY"] = df_soh["TỒN KHO CUỐI NGÀY"].astype(int)
            if  df_soh.loc[ind,"TỒN KHO CUỐI NGÀY"] > 0:  #Hết hạn mà còn tồn kho thì phải HỦY HÀNG
              df_expired = df_expired.append({"NGUYÊN LIỆU":df_soh.loc[ind,"NGUYÊN LIỆU"], "SỐ LƯỢNG": df_soh.loc[ind,"TỒN KHO CUỐI NGÀY"],"ĐƠN VỊ":df_soh.loc[ind,"ĐƠN VỊ"], "NGÀY HẾT HẠN":df_soh.loc[ind,"HẠN SỬ DỤNG"]},ignore_index=True)
            df_soh.loc[ind,"TỒN KHO CUỐI NGÀY"]=0
             
    # Kiểm tra hàng đang về ==> Nếu có đưa qua bảng On-Hand stock
    for ind, data in df_sog.iterrows():
        df_sohh = df_soh
        if df_sog.loc[ind,"NGÀY VỀ"] <= dd: 
            df_soh = df_soh.append({"NGUYÊN LIỆU":df_sog.loc[ind,"NGUYÊN LIỆU"], "TỒN KHO CUỐI NGÀY": df_sog.loc[ind,"SỐ LƯỢNG"],"ĐƠN VỊ":df_sog.loc[ind,"ĐƠN VỊ"], "HẠN SỬ DỤNG":df_sog.loc[ind,"HẠN SỬ DỤNG"]},ignore_index=True)
            df_sohh=df_soh
            df_sog = df_sog.drop(axis = 0, index = ind)
        df_soh = df_sohh
    #Cập nhật tồn kho sau 1 ngày "lao động hăng say"
    df_merge = df_in.merge(df_bom,on = "THÀNH PHẨM", how = 'inner')
    df_merge["NHU CẦU NGUYÊN LIỆU"]=df_merge["SỐ LƯỢNG NGUYÊN LIỆU TRÊN 1 THÀNH PHẨM"]*df_merge["production_qty"]
    df_merge = df_merge.groupby("NGUYÊN LIỆU").sum()
    df_merge = df_soh.merge(df_merge,on="NGUYÊN LIỆU",how = "left")
    df_merge.sort_values(by=["NGUYÊN LIỆU","HẠN SỬ DỤNG"],inplace=True)
    df_merge['inv_cumsum'] = df_merge.groupby("NGUYÊN LIỆU")['TỒN KHO CUỐI NGÀY'].cumsum()
    df_merge['THIẾU'] = df_merge['NHU CẦU NGUYÊN LIỆU'] - df_merge['inv_cumsum']
    df_merge.groupby('NGUYÊN LIỆU').apply(xu_ly_SL_CONLAI)
```
